```python
df = pd.read_csv('netral.csv')
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
def labeling(text, data):
    words = set(text.split())
    kata_kasar = set(data)
    if words & kata_kasar:
        return 'kasar'
    else:
        return 'tidak kasar'
    
df['label'] = df['text'].astype(str).apply(lambda x: labeling(x, data))
df
```

```python
dkasar = pd.DataFrame()
dkasar = df[df['label'] == 'kasar']
dkasar = dkasar.drop('label', axis=1)
dkasar.to_csv('kasar_only.csv')
```

```python
dkasar = pd.DataFrame()
dkasar = df[df['label'] == 'tidak kasar']
dkasar = dkasar.drop('label', axis=1)
dkasar.to_csv('tidak kasar_only.csv')
```

# Low TF-IDF

```python
dsisa = pd.read_csv('tidak kasar_only.csv')
dsisa
```

```python
import json

with open('kata_kasar.json', 'r') as file:
    kamus = json.load(file)

def labeling(text, kamus):
    words = set(text.split())
    kata_kasar = set(kamus.keys())  
    if words & kata_kasar:
        return 'kasar'
    else:
        return 'tidak kasar'
    
dsisa['labeling'] = dsisa['text'].astype(str).apply(lambda x: labeling(x, kamus))
# dsisa = dsisa.drop('Unnamed: 0', axis=1)
dsisa
```

```python
def labeling(text):
    if text == 'kasar':
        return 'negatif'
    else:
        return 'positif'
    
dsisa['sentiment'] = dsisa['labeling'].astype(str).apply(labeling)
# dsisa = dsisa.drop('Unnamed: 0', axis=1)
# dsisa = dsisa.drop('sentimen', axis=1)
dsisa.to_csv('gabungin.csv')
```
